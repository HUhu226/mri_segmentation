```python
import os
import numpy as np
import matplotlib.pyplot as plt

# 定义数据文件夹路径
data_folder = '/code/CZQ/202311/code/10_labels/'

# 定义所有的 label 文件夹
all_label_folders = ['L01_WM', 'L02_GM', 'L03_CSF', 'L05_Scalp', 'L06_EyeBalls',
                     'L07_CompactBone', 'L08_SpongyBone', 'L09_Blood', 'L10_Muscle']

# 创建一个图表以可视化每个标签的 Dice 分数
plt.figure(figsize=(12, 6))

for label_folder in all_label_folders:
    # 构建.npy文件的路径
    result_file = os.path.join(data_folder, label_folder, 'result', 'prediction_record.npy')

    # 加载.npy文件
    result_data = np.load(result_file, allow_pickle=True)

    check_names = result_data[:, 0]
    prediction_results = result_data[:, 1].astype(float)  # 将结果转换为浮点数

    # 创建编号（从0到N-1）作为X轴
    x = list(range(len(check_names)))
    print(x)
    # 绘制折线图
    plt.plot(x, prediction_results, marker='o', linestyle='-', label=label_folder)

plt.title("Line Plot for Checknames")
plt.xlabel(".pth files")
plt.ylabel("Data")
plt.xticks(x, rotation=45)
plt.legend()  # 添加图例
plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

# 定义数据文件夹路径
data_folder = '/tmp/code/brain_seg/10_labels/'

# 定义所有的 label 文件夹
all_label_folders = ['L01_WM', 'L02_GM', 'L03_CSF', 'L05_Scalp', 'L06_EyeBalls',
                     'L07_CompactBone', 'L08_SpongyBone', 'L09_Blood', 'L10_Muscle']
all_label_folders = ['L10_Muscle']

# 创建一个图表以可视化每个标签的 Dice 分数
plt.figure(figsize=(12, 6))

for label_folder in all_label_folders:
    # 构建.npy文件的路径
    result_file = os.path.join(data_folder, label_folder, 'result', 'prediction_record.npy')

    # 检查文件是否存在
    if not os.path.exists(result_file):
        print(f"Skipping {label_folder} as prediction_record.npy doesn't exist.")
        continue

    # 加载.npy文件
    result_data = np.load(result_file, allow_pickle=True)

    check_names = result_data[:, 0]
    prediction_results = result_data[:, 1].astype(float)  # 将结果转换为浮点数

    # 创建编号（从0到N-1）作为X轴
    x = list(range(len(check_names)))
    print(x)
    # 绘制折线图
    plt.plot(x, prediction_results, marker='o', linestyle='-', label=label_folder)

plt.title("Line Plot for Checknames")
plt.xlabel(".pth files")
plt.ylabel("Data")
plt.xticks(x, rotation=45)
plt.legend()  # 添加图例
plt.tight_layout()
plt.show()
```
